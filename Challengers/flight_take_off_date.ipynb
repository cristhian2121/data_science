##### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

#### Load dataset

```python
data_pd = pd.read_csv("M1_final.csv")
data_pd.shape
# Get Tarjet
y = data_pd["TAXI_OUT"]
data_pd = data_pd.drop(columns=["TAXI_OUT"], axis=1)
```

#### Describe Our Data

```python
def exist_null_duplicates(data):
  print(f"** Exists duplicated values? {'Yes' if data.duplicated().sum() > 0 else 'No'}")
  exist_null_data_pd = data.isnull().sum()
  null_data_pd = exist_null_data_pd[exist_null_data_pd > 0]

  print(f"** Exists null values? {'Yes' if len(null_data_pd) > 0 else 'No'}",)
  len(null_data_pd) and print(null_data_pd)
```

```python
def get_types(data):
  columns_data = data.columns
  filter_categoric = data.dtypes == "object"
  filter_float = data.dtypes == "float64"
  filter_int = data.dtypes == "int64"

  print(f"Total columns: {data.dtypes.size}")
  print(f"Total columns type object: {columns_data[filter_categoric].size}")
  print(f"Total columns type int: {columns_data[filter_int].size}")
  print(f"Total columns type float: {columns_data[filter_float].size}")
```

```python
X = data_aux
abs(X.corr())
```

```python
data_pd["Wind"].value_counts()
```

```python
# Start train
X_train, X_test, y_train, y_test = train_test_split(data_aux,y, test_size=0.3, random_state=12, shuffle=True)
```

```python
# Use method
modelo_linear = LinearRegression() 
```

```python
# Train
modelo_linear.fit(X = X_train, y = y_train)
```

```python
# validate
print("Coeficiente de determinación R^2:", modelo_linear.score(data_aux, y))
predicciones = modelo_linear.predict(X = X_test)
rmse = mean_squared_error(
        y_true  = y_test,
        y_pred  = predicciones,
        squared = False
       )
print(f"El error (rmse) de test es: {rmse}")
```
